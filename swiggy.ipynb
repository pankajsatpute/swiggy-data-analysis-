```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("swiggy_clean.csv")
df.head(10)
```

```python
df.describe
```

```python
city_wise_restaurant = df['City'].value_counts().head(10)
plt.Figure(figsize=(10,5))
city_wise_restaurant.plot(kind='bar')
plt.title("top 10 cities restaurant")
plt.xlabel("City")
plt.ylabel("Number of Restaurant")
plt.show()
```

```python
city_wise_price = df.groupby('City')['Price'].mean().sort_values(ascending=False).head(10)

plt.Figure(figsize=(10,5))
city_wise_price.plot(kind='bar')
plt.title("city_wise_avg_price")
plt.xlabel("city")
plt.ylabel("Average price")
plt.show()
```

```python
top_food = df['Food type'].value_counts().head(10)

plt.Figure(figsize=(10,5))
top_food.plot(kind='bar')
plt.title("Top 10 papular Food")
plt.xlabel('Cuisne')
plt.ylabel('count')
plt.show()
```
